**Still getting an error. We dropped all the 'NaN' values. If we are getting error means there are values in 'Order Date' column which are not "date" and something different type but we don't know what are those. How can we find out any invalid values?**

The function **is_valid_date** is designed to check whether a given string can be converted into a valid date. 

Inside the try block, the function attempts to convert date_str to a datetime object using the pandas.to_datetime function.
If date_str can be successfully converted to a datetime object, the function will reach the return True statement, indicating that the input string is a valid date format.

If pd.to_datetime(date_str) raises a ValueError (which happens if the string cannot be parsed as a date), the except block will catch the error and the function will return False. This indicates that the input string is not a valid date format.

```python
# Function to check if a string can be converted to a date
def is_valid_date(date_str):
    try:
        pd.to_datetime(date_str)
        return True
    except ValueError:
        return False
```

~:
The tilde (~) operator is the bitwise NOT operator, which in this context is used to invert the boolean values in the Series returned by apply(is_valid_date).

If apply(is_valid_date) returns True for a valid date, ~ will convert it to False.
If apply(is_valid_date) returns False for an invalid date, ~ will convert it to True.

```python
# Apply the function to the 'Order Date' column and filter invalid dates
invalid_dates = df[~df['Order Date'].apply(is_valid_date)]
```

## Q) What products are most often sold together?

This is a difficult problem. So basically you need to find out that which products sold in pairs. For e.g. iPhone with USB-C Charging cable or any other combination. Approach of this probelm in SQL and in Python is very different. 

```python
df_cleaned.head()
```

**So to find out the products sold togather, we need to look at the 'Order ID'. Because You can buy many product in one order which means if someone buys 2 or more products, it will have same 'Order ID' which means we can find the pair which were sold most by counting the all pairs and then get the max values of that count. For that we somehow need to form a pair**

```python
df = df_cleaned[df_cleaned['Order ID'].duplicated(keep=False)]
df
```

```python
df['groupped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df
```

```python
## dropping the duplicated values.

df = df[['Order ID','groupped']].drop_duplicates()
df.head(20)
```

### To create combinations of products we need to import different library. I used reference of stackoverflow. 

reference:-  https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for i in df['groupped']:
    i_list = i.split(',')
    count.update(Counter(combinations(i_list,2)))

for key, value in count.most_common(10):
    print(key,value)
```

### Explanation of above code in detail 

Detailed Explanation Libraries

itertools:
This is a standard Python library that provides functions for creating iterators for efficient looping.
combinations is a function from itertools that returns all possible combinations of a specified length from the input iterable.

collections:
This is another standard Python library that provides specialized container datatypes.
Counter is a class from collections that is used to count hashable objects. It is a dictionary subclass designed to count the frequency of elements.

**combinations will be used to generate all possible pairs of products.**

**Counter will be used to count the frequency of each product pair.**

count = Counter() :-  A Counter object named count is initialized to keep track of the frequency of product pairs.




    1) combinations(i_list, 2):-  Generates all possible pairs of products (combinations of length 2) from the list i_list.

    2) Counter(combinations(i_list, 2)):- Creates a temporary Counter object to count each product pair in the current list.

    3) count.update(...):- Updates the main count Counter with the counts from the temporary Counter.
